| 9     | Minuto luz a kilómetro               | 17987539.67|

Para este ejercicio se deben almacenar los factores de conversión en un vector (tipo de dato ```list``` de Python) para luego:

- Solicitar al usuario el dato que quiere convertir para que lo ingrese por teclado.
- Tranformar el dato tipo texto ```str``` en tipo numérico (```int``` o ```float``` según sea el caso).
- Acceder al factor de conversión correspondiente de acuerdo a su posición o índice en el vector de factores de conversión.


````python
import math
import numpy

# Define the conversion factors correctly
factores = [1000, 0.01, 0.62, 20, 0.05, 1, 4200, 4200, 0.45, 17987539.67]

# Prompt the user to select the factor for conversion
Factor = int(input("Ingrese el índice del factor al que desea convertir (0-9): "))

# Check if the index is within the valid range of the list
if 0 <= Factor < len(factores):
    # Prompt the user to enter the value they want to convert
    ValorConvertir = float(input("Ingresa el valor que desea convertir: "))

    # Perform the conversion
    ValorConvertido = ValorConvertir * factores[Factor]
    unidad = [" metros", " metros", " millas por hora", " nota en canvas", " nota en educatic", " centimetro cubico", " peso colombiano"," dolar", " kilometro", " kilometro"]
    unidad2 = [" kilometro", " centimetros", " kilometros", " nota en educatic", " nota en canvas", " milimetro", " dolares"," pesos", " libras", " minutos luz"]
    kill = unidad[Factor]
    dead = unidad2[Factor]

    # Display the final value
    print(f"{ValorConvertir, dead} equivalen a {ValorConvertido, kill}")

else:
    print("Índice no válido. Por favor, elija un índice entre 0 y 9.")
````

# (25%) Ejercicio 2

La distancia entre dos puntos en un plano se calcula con la siguiente fórmula:

> $d(P_1, P_2)=\sqrt{{(x_2-x_1)}^2+{(y_2-y_1)}^2}$

Donde las coordenadas *x* y *y* indican la posición en la que se encuentra cada punto *P*.

Realiza un algoritmo que permita al usuario ingresar los cuatro valores de las coordenadas en orden (debes indicarle cuál coordenada está ingresando) y calcular la distancia *d*.

El resultado debe presentar dos cifras decimales y separador de miles.

Además, para el presente ejercicio necesitas importar la librería matemática ```math``` y usar sus componentes de potenciación y radicación.

````python

import math
import numpy
x1 = int(input("ingrese el valor de x1: " ))
x2 = int(input("ingrese el valar de x2: " ))
y1 = int(input("ingrese el valar de x1: " ))
y2 = int(input("ingrese el valar de y2: " ))

d = math.sqrt((x2-x1)**2 + (y2-y1)**2)

print("la distancia entre los puntos es: ", d)
````

¿Qué es el currículum vitae?
El currículum vitae (CV) es un resumen breve de tus experiencias formativas y laborales y de tus habilidades profesionales.

Su objetivo es demostrar la idoneidad de tu candidatura a un puesto de trabajo concreto y permitirte acceder al proceso de selección.

El CV es una herramienta esencial para el desarrollo profesional de cualquier persona. Conviene tenerlo siempre actualizado, pues nunca sabes en qué momento de tu carrera se puede presentar la ocasión de optar al puesto deseado.

Elementos del currículum
Los principales elementos o apartados que se recomienda incluir en el CV son los siguientes:



Nombre y apellidos + profesión
Tu nombre va asociado a tu marca personal. Conviene que aparezca en un lugar destacado del currículum, al principio de este, con un formato atractivo.

Bajo el nombre se recomienda citar la profesión o cargo que deseas desarrollar en el puesto al que optas.

Dicha profesión deberá tener conexión con la trayectoria y formación que indiques en el CV.



Fotografía
Es recomendable que incluyas tu fotografía al inicio del CV, cerca de tu nombre.

Asegúrate de que la calidad de la imagen es buena, tiene que verse correctamente.

Elige una foto reciente. Si pasas a las siguientes fases del proceso de selección conviene que la imagen del CV responda a tu imagen actual.

En algunos sectores profesionales conviene que la fotografía sea formal. Aunque no sea tu caso, intenta transmitir siempre una imagen positiva a través de la foto que insertes en el CV.



Datos personales
Es importante que incluyas el teléfono y correo electrónico para que la empresa pueda contactar contigo.

También puede ser de utilidad enlazar a tus redes sociales profesionales, página web, blog, etc. De este modo, los responsables del proceso de selección podrán ampliar la información relativa a tu trayectoria e intereses.

Aunque no es imprescindible indicar tu dirección completa, puedes incluir tu provincia o ciudad de residencia como información complementaria.



Perfil profesional
Es un breve resumen de las habilidades y logros profesionales que desees destacar para dejar constancia de tus expectativas laborales y de lo que puedes aportar a la empresa.

El perfil puede incluir experiencias personales o valores que contribuyan a reforzar la marca personal.

Este apartado puede titularse simplemente "Perfil", "Quién soy", "Sobre mí", o incluso obviarse el título, pues se identifica fácilmente qué información contiene.

Se recomienda ubicar el perfil al inicio del CV, después de la fotografía o del nombre y el cargo.



Experiencia laboral
Este apartado debe incluir los empleos realizados anteriormente.

Para cada empleo debes indicar:

Nombre de la empresa y ciudad en la que está ubicada.
Inicio y finalización de contrato (mes y año).
Cargo ocupado.
Funciones desempeñadas.
Puedes añadir la experiencia laboral sin contrato y, si todavía no tienes experiencia laboral o esta es muy escasa, las prácticas o actividades de voluntariado realizadas.

Utiliza el orden cronológico inverso, es decir, escribe en primer lugar tu actual puesto de trabajo o el más reciente.



Formación académica
Este apartado recoge los estudios oficiales que hayas cursado. Conviene señalar el año en el que obtuviste cada titulación y el centro donde cursaste los estudios.

Indica tu formación por orden cronológico inverso, escribiendo en primer lugar la titulación académica más reciente.



Formación complementaria
Se trata de un apartado opcional en el que se pueden detallar los cursos de formación no reglada, jornadas, seminarios, etc. que te interese reflejar.

Evita incluir información innecesaria, no hace falta añadir formación adicional que no tenga relación con el trabajo que estás buscando.



Competencias/Habilidades
En este apartado debes incluir las competencias profesionales que has desarrollado a lo largo de tu trayectoria académica y laboral.

Conviene que distingas entre:

Competencias técnicas. Son conocimientos específicos, herramientas y procedimientos asociados a ocupaciones concretas. Puedes indicar, por ejemplo, tu dominio de lenguajes y programas informáticos especificando el nivel.
Competencias transversales, también conocidas como soft skills. Son capacidades transferibles a cualquier entorno laboral, como pueden ser el trabajo en equipo o el liderazgo.


Idiomas
En este apartado deberás citar los idiomas que conoces.

Debes especificar el nivel de conocimiento de cada idioma, la titulación que lo acredita (por ejemplo, First Certificate) y el año en el que la obtuviste.

En caso de que no tengas ningún título, indica el nivel de conocimiento para cada idioma: básico, intermedio, avanzado o nativo.



Otros datos de interés
Aquí se incluye información que no encaja en los apartados anteriores, pero que puede ser útil para los responsables de la selección.

Por ejemplo, puedes indicar si tienes permiso de conducir y vehículo propio, tu disponibilidad horaria, cuándo podrías incorporarte, si tienes cartas de recomendación, aficiones e intereses (sobre todo si tienen relación con el puesto de trabajo).
